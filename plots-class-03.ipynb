```python
import matplotlib.pyplot as plt

legoGroth16ProofTimes = [ 19.444, 0.049, 14.284, 28.596, 42.77 ]
bulletProofProofTimes = [ 0.199, 0.05, 0.671, 1.277, 1.913 ]

# Create the bar chart
plt.bar(
    [i - 0.1 for i in range(len(legoGroth16ProofTimes))],
    legoGroth16ProofTimes,
    width=0.2,
    label="LegoGroth16",
    color="blue",
)
plt.bar(
    [i + 0.1 for i in range(len(bulletProofProofTimes))],
    bulletProofProofTimes,
    width=0.2,
    label="BulletProof++",
    color="red",
)

# Add labels and title
plt.xlabel("Number of range proofs")
plt.ylabel("Proof Generation Time (seconds)")
plt.title("Comparison of Proof Generation Times")
plt.xticks(range(len(legoGroth16ProofTimes)), ["Setup", "0", "1", "2", "3"])
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

legoGroth16Size = [0.491, 0.892, 1.293, 1.694]
bulletProofSize = [0.491, 1.798, 3.105, 4.412]

# Create the bar chart
plt.bar(
    [i - 0.1 for i in range(len(legoGroth16Size))],
    legoGroth16Size,
    width=0.2,
    label="LegoGroth16",
    color="blue",
)
plt.bar(
    [i + 0.1 for i in range(len(bulletProofSize))],
    bulletProofSize,
    width=0.2,
    label="BulletProof++",
    color="red",
)

# Add labels and title
plt.xlabel("Number of range proofs")
plt.ylabel("Size of proof in K.bytes")
plt.title("Comparison of Proof Sizes")
plt.xticks(range(len(legoGroth16Size)), ["None", "1", "2", "3"])
plt.legend()

# Show the plot
plt.show()
```
