## Proof Generation Times Comparison

```python
import matplotlib.pyplot as plt

legoGroth16ProofTimes = [10.563, 0.027, 5.954, 11.906, 17.98] 
bulletProofProofTimes = [0.113, 0.028, 0.36, 0.692, 1.051] 

# Create the bar chart
plt.bar([i - 0.1 for i in range(len(legoGroth16ProofTimes))], legoGroth16ProofTimes, width=0.2, label="LegoGroth16", color="blue")
plt.bar([i + 0.1 for i in range(len(bulletProofProofTimes))], bulletProofProofTimes, width=0.2, label="BulletProof++", color="red")

# Add labels and title
plt.xlabel("Number of range proofs")
plt.ylabel("Proof Generation Time (seconds)")
plt.title("Comparison of Proof Generation Times")
plt.xticks(range(len(legoGroth16ProofTimes)), ["Setup", "0", "1", "2", "3"])
plt.legend()

# Show the plot
plt.show()
```

## Server load -- estimated through cumulative processing time
A - graph should show total verifier time-based load for LG16 (blue, setup + verification) and Bp++ (red, setup + verification)  
B - graph should show total holder time-based load for LG16 (blue, proof creation) and Bp++ (red, setup + proof creation)

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['0', '1', '2', '3']

# Total verifier time-based load for LG16 (blue, setup + verification) and Bp++ (red, setup + verification)  
lg16_verifiers_setup_times = [0, 10.563, 10.563, 10.563]
lg16_verifiers_verification_times = [0.031, 0.133, 0.235, 0.342]

bp_verifiers_setup_times = [0.113, 0.113, 0.113, 0.113]
bp_verifiers_verification_times = [0.03, 0.048, 0.06, 0.073]

# Total holder time-based load for LG16 (blue, proof creation) and Bp++ (red, setup + proof creation)
lg16_holders_setup_times = [0, 0, 0, 0]
lg16_holders_proof_creation_times = [0.027, 5.954, 11.906, 17.98]

bp_holders_setup_times = [0.113, 0.113, 0.113, 0.113]
bp_holders_proof_creation_times = [0.028, 0.36, 0.692, 1.051]

width = 0.2       # the width of the bars: can also be len(x) sequence

fig, axes = plt.subplots(1, 2, figsize=[12, 5])

# ------------------------------ Setting up first graph For verifiers -------------------------------------------------------------------------

axes[0].bar([i - 0.11 for i in range(4)], lg16_verifiers_setup_times, width, label='LG16 setupTimes', color="blue")
axes[0].bar([i - 0.11 for i in range(4)], lg16_verifiers_verification_times, width, bottom=lg16_verifiers_setup_times, label='LG16 Proof Generation Times', color="orange")

axes[0].bar([i + 0.11 for i in range(4)], bp_verifiers_setup_times, width, label='BP++ setupTimes', color="red")
axes[0].bar([i + 0.11 for i in range(4)], bp_verifiers_verification_times, width, bottom=bp_verifiers_setup_times, label='BP++ Proof Generation Times', color="purple")

axes[0].set_ylabel('Proof Verification Time (seconds)')
axes[0].set_title('Total verifier load (based on total consumed time)')
plt.xticks(range(4), ["0", "1", "2", "3"])
axes[0].legend()

# ------------------------------ Setting up Second graph for holders -------------------------------------------------------------------------

axes[1].bar([i - 0.11 for i in range(4)], lg16_holders_setup_times, width, label='LG16 setupTimes', color="blue")
axes[1].bar([i - 0.11 for i in range(4)], lg16_holders_proof_creation_times, width, bottom=lg16_holders_setup_times, label='LG16 Proof Generation Times', color="orange")

axes[1].bar([i + 0.11 for i in range(4)], bp_holders_setup_times, width, label='BP++ setupTimes', color="red")
axes[1].bar([i + 0.11 for i in range(4)], bp_holders_proof_creation_times, width, bottom=bp_holders_setup_times, label='BP++ Proof Generation Times', color="purple")

axes[1].set_ylabel('Proof Generation Time (seconds)')
axes[1].set_title('Total holder load (based on total consumed time)')
plt.xticks(range(4), ["0", "1", "2", "3"])
axes[1].legend()


plt.show()
```
